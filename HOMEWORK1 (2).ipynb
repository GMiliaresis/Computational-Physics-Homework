```python

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math
```

1.1 Just do one big experiment

```python
dimension=2
X = 1
P = 10000
points = np.random.uniform(low=-1.0, high=1.0,
size=(dimension, P,X, ))
radius = (points**2).sum(axis=0)
inside_the_circle = (radius <= 1)
outside_the_circle = (radius > 1)
inside_the_square = ((-1.0 <= points[0]) & (points[0] <= 1.0) &
(-1.0 <= points[1]) & (points[1] <= 1.0))
inside=[]
for i in range(10000):
    x= 4 if inside_the_circle[i]==True else 0
    inside.append(x)

pi=sum(inside)/P
s=[x-pi for x in inside]

ssquared=[x**2 for x in s]
er=math.sqrt(1/(P)*sum(ssquared))
print("The estimate is π =",pi,"+-",er)

```

```python

plt.hist(inside,20)
plt.axvline(x=pi, color='red', linestyle='--',label="mean")
plt.axvline(x=pi-er, color='yellow', linestyle='--',label="uncertainty")
plt.axvline(x=pi+er, color='yellow', linestyle='--')
plt.axvline(x=math.pi, color='black', linestyle='-',label="pi")
plt.title("1.1 Just do one big experiment")
leg = plt.legend(loc='upper left')
```

1.2 Split into 100 experiments

```python
dimension=2
P=100 
X=100
points = np.random.uniform(low=-1.0, high=1.0,
size=(dimension, P,X, ))
radius = (points**2).sum(axis=0)
inside_the_circle = (radius <= 1)
inside_the_square = ((-1.0 <= points[0]) & (points[0] <= 1.0) &
(-1.0 <= points[1]) & (points[1] <= 1.0))
pi_by_4 = inside_the_circle.sum(axis=0) / inside_the_square.sum(axis=0)
pi = 4 * pi_by_4
avg=sum(pi)/X
s=pi-avg
ssquared=s**2
er=math.sqrt(1/(X)*sum(ssquared))
print("The mean is pi=",avg)
print("The uncetainty =",er)
plt.hist(pi,15)
plt.axvline(x=avg, color='red', linestyle='--',label="mean")
plt.axvline(x=avg-er, color='yellow', linestyle='--',label="uncertainty")
plt.axvline(x=avg+er, color='yellow', linestyle='--')
plt.axvline(x=math.pi, color='black', linestyle='-',label="pi")
plt.title("1.2 Split into 100 experiments")
leg = plt.legend(loc='upper left')
```

In all of the cases the the estimate for pi is compatible with the real value.
The use of the standard deviation does not make sense because the  outcomes of the measurements do not follow a normal distribution but have only two values 0(out) and 4 (in).
We can see that the uncertainty decreases when we have around the same number of experiments and pairs. In this case we have a big enough number of pairs to take a rough estimate π_x (not just 0 or 4 like in P=1 case) and we average over a big enough number of experiments. When X is big enough the distribution of π_x is close to the normal distribution because of the central limit theorem.

1.5 More Experiments vs. Longer Experiments

```python
average=[]
error=[]
def experiment(X,P):
    points = np.random.uniform(low=-1.0, high=1.0,
    size=(dimension, P,X, ))
    radius = (points**2).sum(axis=0)
    inside_the_circle = (radius <= 1)
    inside_the_square = ((-1.0 <= points[0]) & (points[0] <= 1.0) &
    (-1.0 <= points[1]) & (points[1] <= 1.0))
    pi_by_4 = inside_the_circle.sum(axis=0) / inside_the_square.sum(axis=0)
    pi = 4 * pi_by_4
    avg=sum(pi)/X
    s=pi-avg
    ssquared=s**2
    er=math.sqrt(1/(X)*sum(ssquared))
    average.append(avg)
    error.append(er)
    print("Case X=",X," and P=",P)
    print("The mean is pi=",avg)
    print("The uncertainty =",er)
    plt.hist(pi,15)
    plt.axvline(x=avg, color='red', linestyle='--',label="mean")
    plt.axvline(x=avg-er, color='yellow', linestyle='--',label="uncertainty")
    plt.axvline(x=avg+er, color='yellow', linestyle='--')
    plt.axvline(x=math.pi, color='black', linestyle='-',label="pi")
    leg = plt.legend(loc='upper left')
    plt.show()
    return pi
Pvalues= [10,100,1000,10000]
Xvalues= [10,100,1000,10000]
for X in Xvalues:
    for P in Pvalues:
        experiment(X,P)
```
